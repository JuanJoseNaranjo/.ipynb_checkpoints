### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
```

```python
chipo.head(10)
```

```python
chipo['item_price'].dtype
```

```python
duplicates = chipo.duplicated()
duplicates.value_counts()
```

### Step 4. How many products cost more than $10.00?

```python
# Transforming item_price into a float to manipulate it
chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float)
```

```python
expensive_items = chipo[chipo['item_price'] > 10.00]
expensive_items
```

```python
# deleting duplicate values in item_name, quantity and choice_description
chipo_without_duplicate = chipo.drop_duplicates(['item_name','quantity','choice_description'])
chipo_without_duplicate.duplicated().value_counts()
```

```python
# select only the products with quantity equals to 1
chipo_unique = chipo_without_duplicate[chipo_without_duplicate.quantity == 1]
chipo_unique
```

```python
bigger_than_10 = chipo_unique[chipo_unique['item_price'] > 10.00]
bigger_than_10
```

```python
bigger_than_10.shape[0]
```

### Step 5. What is the price of each item?
###### print a data frame with only two columns item_name and item_price

```python
price_per_item = chipo_unique[['item_name', 'item_price']]
price_per_item
```

```python
price_per_item = price_per_item.drop_duplicates(['item_name','item_price'])
price_per_item
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values("item_name")
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo.sort_values(by = "item_price", ascending = False).head(1)
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
len(chipo[chipo.item_name == "Veggie Salad Bowl"])
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
len(chipo[(chipo.item_name == "Canned Soda") & (chipo.quantity > 1)])
```
